```python
df = df.dropna()
df
```

```python
df = df.set_index(['State', 'Year'])
df.head()
```

```python
from linearmodels import PanelOLS

Y = df['gini']
df['const'] = 1
dd = ['const', 'Normalized PL05', 'Time']
dif_in_dif = PanelOLS(Y, df[dd]).fit(cov_type = 'clustered', cluster_entity = True)
```

```python
firmFE = PanelOLS(Y, df[dd], entity_effects = True)
print(firmFE.fit(cov_type = 'clustered', cluster_entity = True))
```

```python
stateFE = PanelOLS(Y, df[dd + ['Abb']])
print(stateFE.fit(cov_type = 'clustered', cluster_entity = True))
```
