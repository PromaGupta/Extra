```python
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])

plt.plot(a)
plt.plot(b)
plt.show()


## Note: You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or
array to plot, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you.
Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the 
x data are [0, 1, 2, 3].
```

```python
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])
c = np.array([3,2,4,6,5])

plt.plot(a, c = 'green', linestyle = '--', label = "Rent Y20", linewidth = 2)
plt.plot(b, c = 'red', linestyle = '-.', label = "Rent Y21", linewidth = 4)
plt.plot(c, c = 'black', label = "Rent Y22", linewidth = 5)
plt.title("Rent per year", fontsize = 30, fontweight = "bold")
plt.legend()
plt.show()
```

```python
##xticks

a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])
c = np.array([3,2,4,6,5])

plt.plot(a, c = 'green', linestyle = '--', label = "Rent Y20", linewidth = 2)
plt.plot(b, c = 'red', linestyle = '-.', label = "Rent Y21", linewidth = 4)
plt.plot(c, c = 'black', label = "Rent Y22", linewidth = 5)
plt.title("Rent per year", fontsize = 30, fontweight = "bold")
plt.legend(loc = "upper right", fontsize = 15)
plt.xticks(range(0,len(a)),('Jan', 'Feb', 'Mar', 'Apr', 'May'))
plt.show()
```
